### Importing Libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
import pickle
```

### Data Preprocessing

```python
# In the dataset, the Id column is irrelevent and having no relationship with outcome , so removing it.
df.drop(["ID"],axis =1, inplace=True)
```

```python
df.head()
```

```python
# checking the missing values
df.isnull().sum()
```

```python
# checking the datatype
df.info()
```

```python
df.duplicated().sum()
```

#### so no duplicated data

```python
# Handling Gender column
df["Gender"].unique()
```

```python
df["Gender"].value_counts()
```

##### Here we can see that - and # are the missing values and we will remove it using the most frequent data.

```python
most_frequent = df["Gender"].mode()[0]
most_frequent
```

```python
# Replacing the missing values with the most frequent data "M"
df["Gender"] = df["Gender"].replace(["-" , "#" ], most_frequent)
```

```python
df.isnull().sum()
```

##### Hence the columns Gender, Home Ownership, Income ,and online contains mising values so we are using simpleImputer to handle the missing values.

```python
# Here are 2 types of columns , one is categorical and another is numerical.
categorical_imputer = SimpleImputer(strategy="most_frequent")
numerical_imputer = SimpleImputer(strategy="median")
```

```python
df[["Gender","Home Ownership" ]] = categorical_imputer.fit_transform(df[["Gender","Home Ownership"]])
df[["Income" , "Online"]] = numerical_imputer.fit_transform(df[["Income" , "Online"]])
```

```python
df.isnull().sum()
```

```python
#checking columns manually
for column in df.columns:
    print(f"Column: {column}")
    print(df[column].value_counts())
    print()
```

```python
# here we can see that the column personal loan contains one missing values so handling it
df["Personal Loan"] = df["Personal Loan"].replace(" " , "1").astype(float)
```

```python
df["Personal Loan"].value_counts()
```

```python
df["Age"].value_counts()
```

##### Here we can see that the ages like 0,2 ,4,7,8,600, 786 etc are the outliers so removing this

```python
# Defining a range of valid range for loan
valid_age_range = (20, 100)

# Filtering the DataFrame to include only valid ages
valid_ages = df[(df['Age'] >= valid_age_range[0]) & (df['Age'] <= valid_age_range[1])]['Age']

# Calculating the median age within the valid range
median_age = valid_ages.median()

# Identifying outliers 
outliers = df[(df['Age'] < valid_age_range[0]) | (df['Age'] > valid_age_range[1])]

# Replacing the outliers with the median age
df.loc[outliers.index, 'Age'] = median_age
```

```python
df["Age"].value_counts().sort_index()
```

```python
df["Experience"].value_counts()
```

##### Removing negative experience values

```python
df = df[df["Experience"]>0]
```

```python
df["Experience"].value_counts()
```

### Feature Transformation

```python
# Separating the dataset into input features (X) and target variable (y)
X= df.drop(columns=["Personal Loan"],axis=1)
y= df["Personal Loan"]
```

```python
# Assuming you have defined numerical_features and categorical_features earlier
numerical_features = X.select_dtypes(exclude='object').columns
categorical_features = X.select_dtypes(include='object').columns

# Define transformers for numerical and categorical features
num_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot_encoder', OneHotEncoder())
])

# Create a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, numerical_features),
        ('cat', cat_transformer, categorical_features)
    ]
)
```

```python
# Preprocess the data
X_processed = preprocessor.fit_transform(X)
```

```python
# Split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Define a dictionary of models
models = {
    "Logistic Regression": LogisticRegression(),
    "K-Neighbors Classifier": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest Classifier": RandomForestClassifier(),
    "AdaBoost Classifier": AdaBoostClassifier(),
    "XGBClassifier": XGBClassifier(),
    "Naive Bayes": GaussianNB(),
}
```

```python
# Evaluating model

for name, model in models.items():
    model.fit(x_train, y_train)
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    
    # Calculating evaluation metrics
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred)
    recall = recall_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred)
    confusion = confusion_matrix(y_test, y_test_pred)

    
     # Printing results
    print(f"Model: {name}")
    print("Training Accuracy:", train_accuracy)
    print("Test Accuracy:", test_accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)
    print("Confusion Matrix:")
    print(confusion)
    print("\n")
```

## I selected the Random Forest Classifier as the best model due to its better performance compared to other algorithms. It achieved the highest test accuracy (98.36%), precision (97.59%), recall (85.26%), and F1 score (91.01%), making it the most reliable choice for predicting Personal Loan approval.

#### Dumping the model

```python

random_forest_model = RandomForestClassifier()
random_forest_model.fit(x_train, y_train)

# defining the filename
filename = "random_forest_model.pickle"

# Saving model using pickle
with open(filename, 'wb') as file:
    pickle.dump(random_forest_model, file)
```

```python
# Loading the mode
loaded_model = pickle.load(open(filename, "rb"))

# Use the loaded model to make predictions on the training data
y_predicted = loaded_model.predict(x_train)
```

```python
# Calculating evaluation metrics
accuracy = accuracy_score(y_train, y_predicted)
precision = precision_score(y_train, y_predicted)
recall = recall_score(y_train, y_predicted)
f1 = f1_score(y_train, y_predicted)

# printing Evaluation Metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
